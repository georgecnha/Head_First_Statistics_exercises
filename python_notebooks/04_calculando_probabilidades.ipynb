# 4 - Calculando Probabilidades

A probabilidade é uma forma de medir as chances de algo acontecer. Ela é medida em uma escala de 0 a 1. Se um evento é impossível, ele tem probabilidade 0. Se for um evento com certeza absoluta, então a probabilidade é 1.

$$\textbf{P(A)}= \dfrac{\textbf{n(A)}}{\textbf{n(S)}}$$

- P(A): Probabilidade de ocorrência do evento A.
- n(A): Número de formas de obter um evento A.
- n(S): Número de resultados possíveis.

**S** é conhecido como **espaço de possibilidades**, ou **espaço amostral**. É uma forma resumida de se referir a todos os resultados possíveis. Todos os eventos possíveis são subconjuntos de S. 

### Eventos Complementares

Há uma forma resumida de indicar o evento em que A não ocorre -- A'. A' é conhecido como o evento complementar de A.

Se algo está em A, ele não pode estar em A' e, se algo não está em A, ele deve estar em A'. Logo P(A)+P(A')=1.

$$\textrm{P(A) + P(A') = 1 }$$ ou

$$\textbf{P(A') = 1 - P(A) }$$

4. Finally, use your roulette board to count all the holes that are either black or even, then divide by the total number of holes. What do you get?

____________________________________

Se dois eventos são **mutuamente exclusivos**, apenas um dos dois pode ocorrer a cada vez.




Se dois eventos se interceptam, eles são **não-exclusivos**. É possível que ocorram simultaneamente.

____

Calculating the probability of getting a black or even went wrong because we included black and even pockets twice.
Here’s what happened. First of all, we found the probability of getting a black pocket and the probability of getting an even number. 

When we added the two probabilities together, we counted the probability of getting a black and even pocket twice.


To get the correct answer, we need to subtract the probability of getting both black and even. This gives us:

$\text{P(Black or Even) = P(Black) + P(Even) - P(Black and Even)}$

$\text{P(Black or Even) =} \frac{18}{38} + \frac{18}{38} - \frac{10}{38} = \frac{26}{38} = 0.684$

____



Calcular a probabilidade de obter preto ou par deu errado porque as divisórias que são simultaneamente pretas e pares foram incluídas duas vezes. O que aconteceu foi o seguinte: Primeiramente nós encontramos a probabilidades de obter uma divisória preta e a probabilidade de obter um número  par (separadamente). 

Quando nós somamos as duas probabilidades, contamos a probabilidade de obter uma divisória preta **e** par duas vezes.

Para obter a resposta correta, precisamos subtrair a probabilidade de obter simultaneamente preto **e** par. Isto resulta em:

$\text{P(Preto ou Par) = P(Preto) + P(Par) - P(Preto e Par)}$

$\text{P(Preto ou Par) =} \frac{18}{38} + \frac{18}{38} - \frac{10}{38} = \frac{26}{38} = 0.684$

There’s a more general way of writing this using some more math shorthand.

First of all, we can use the notation A $\cap$ B to refer to the intersection between A and B. You can think of this symbol as meaning “and.” It takes the common elements of events

$A \cup B$, on the other hand, means the union of A and B. It includes all of the elements in A and also those in B.
You can think of it as meaning “or.” If $A \cup B =1$, then A and B are said to be **exhaustive**.

Between them, they make up the whole of S. They exhaust all possibilities.

____

Existe um jeito mais genérico de escrever isso usando simbologia matemática. 

Pode-se utilizar a notação A $\cap$ B para referir-se à interseção entre A e B. Pode-se dizer que este símbolo significa "e". Isto inclui os elementos em comum entre os eventos.

$A \cup B$, por outro lado, significa a união entre A e B. Isto inclui todos os elementos em A e também aqueles em B. Pode-se dizer que este símbolo significa "ou." Se $A \cup B = 1$, então A e B são chamados **eventos exaustivos**.

Entre si, eles compoem todo o S. Eles esgotam todas as possibilidades.

If we want to express the probability of one event happening given another one has already happened, we use the “|” symbol
to mean “given.” Instead of saying “the probability of event A occurring given event B,” we can shorten it to say P(A | B). So now we need a general way of calculating P(A | B). What we’re interested in is the number of outcomes where both A and B occur, divided by all the B outcomes. Looking at the Venn diagram, we get:

$\textbf{P (A | B) =} \dfrac{P(A \cap B)}{P(B)} $

_____

Se quisermos expressar a probabilidade de um evento acontecer dado que outro já tenha acontecido, usamos o símbolo "|"
para significar "dado". Em vez de dizer "a probabilidade de o evento A ocorrer dado o evento B", podemos encurtá-lo para dizer P(A | B). Então, agora precisamos de uma maneira geral de calcular P(A | B). O que nos interessa é o número de resultados em que A e B ocorrem, dividido por todos os resultados B. Olhando para o diagrama de Venn, obtemos:


$\textbf{P (A | B) =} \dfrac{P(A \cap B)}{P(B)} $
