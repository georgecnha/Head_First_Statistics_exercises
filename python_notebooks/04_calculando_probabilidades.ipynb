# 4 - Calculando Probabilidades

A probabilidade é uma forma de medir as chances de algo acontecer. Ela é medida em uma escala de 0 a 1. Se um evento é impossível, ele tem probabilidade 0. Se for um evento com certeza absoluta, então a probabilidade é 1.

$$\textbf{P(A)}= \dfrac{\textbf{n(A)}}{\textbf{n(S)}}$$

- P(A): Probabilidade de ocorrência do evento A.
- n(A): Número de formas de obter um evento A.
- n(S): Número de resultados possíveis.

**S** é conhecido como **espaço de possibilidades**, ou **espaço amostral**. É uma forma resumida de se referir a todos os resultados possíveis. Todos os eventos possíveis são subconjuntos de S. 

### Eventos Complementares

Há uma forma resumida de indicar o evento em que A não ocorre -- A'. A' é conhecido como o evento complementar de A.

Se algo está em A, ele não pode estar em A' e, se algo não está em A, ele deve estar em A'. Logo P(A)+P(A')=1.

$$\textrm{P(A) + P(A') = 1 }$$ ou

$$\textbf{P(A') = 1 - P(A) }$$

### **Exercício p.137**

###  BE the croupier

#### Your job is to imagine you're the croupier and work out the probabilities of various events. For each event below, write down the probability of a successful outcome.


- P(9)

- P(Verde)

- P(Preto)

- P(38)

```python
"""P(9)"""

#Só existe 1 número 9 em 38 casas.

P09= 1/38

print("P(9) = ", round(P09,3))
```

```python
"""P(Verde)"""

#Só existem 2 número verdes (0 e 00) em 38 casas.

PV= 2/38

print("P(Verde) = ", round(PV,3))
```

```python
"""P(Preto)"""

#38 casas - 2 verdes = 36 casas = 18 pretas e 18 cinzas


PPreto = 18/38

print("P(Preto) = ", round(PPreto,3))
```

```python
"""P(38)"""

#Não existe casa de número 38 em 38 casas.

P38 = 0/38

print("P(Preto) = ", round(P38,3))
```
